Fraud Detection POC

Import packages

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading Data and initial data cleaning

```python
# Load dataset
df = pd.read_csv('fraud.csv',quotechar="'")
# Get rid of variables with null variance and the ID
df.drop(['zipcodeOri','zipMerchant','customer'],axis=1,inplace=True)
print(df.columns)
#Checking if any null values
null_values = df.isnull().sum() ## Result = 0 - Ignore handling null values
#pd.DataFrame(null_values)
```

Variable Analysis

Clean-up unusual records

```python
# Distribution of target variable
sns.countplot(x='fraud', data=df)
plt.show()
# The target output is unbalanced
# Distribution of the qualitative regressors
sns.countplot(x='age', data=df)
plt.show()
sns.countplot(x='gender', data=df)
plt.show()
sns.countplot(x='merchant', data=df)
plt.show()
sns.countplot(x='category', data=df)
plt.show()
# We have to clean up the unusual records
# Pairplot for feature relationships
#sns.pairplot(df, hue='fraud')
#plt.show()
# There's a correlation between the amount and the likelihood that a fraud occurred
```

```python
df.groupby(['age','gender']).size().unstack(fill_value=0)
df.loc[df['age']=='U'].shape[0]/df.shape[0]
df.loc[df['gender']=='U'].shape[0]/df.shape[0]
# Only the 0.1% of records have the "U" and E - I'll remove them
df.drop(df[df.age == 'U'].index,inplace=True)
df.drop(df[df.gender == 'U'].index,inplace=True)
df['category'] = df['category'].apply(lambda x: 'es_transportation' if x == 'es_transportation' else 'other')
df['merchant'] = df['merchant'].apply(lambda x: x if x in ['M1823072687', 'M348934600','M85975013'] else 'other')
```

One hot encoding

```python
data = pd.get_dummies(df, columns=['age', 'gender','category','merchant'])
```

Rescale data and deal with undersampling

```python
X = data.drop(columns=['fraud'])
y = data['fraud']
scaler = StandardScaler()
X[['amount', 'step']] = scaler.fit_transform(X[['amount', 'step']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=71195)
```

```python
rf = RandomForestClassifier(random_state=42)
gb = GradientBoostingClassifier(random_state=42)
lr = LogisticRegression(random_state=42)
nn = MLPClassifier(random_state=42)

models = [rf, gb, lr, nn]
for model in models:
    model.fit(X_train, y_train)
```

Metrics

```python
models = {'RandomForest': rf, 'LogisticRegression': lr, 'GradientBoosting': gb, 'NeuralNetwork': nn}
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f'{name} F1 Score: {f1_score(y_test, y_pred)}')
    print(f'{name} AUC-ROC: {roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])}')
```

Model Interpretation

```python
importances = rf.feature_importances_
feature_importance = pd.Series(importances, index=X.columns).sort_values(ascending=False)
a = pd.DataFrame(feature_importance)
a
```
